Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
```

Step 2 — Load Both CSV Files

```python
# Load datasets
mat_df = pd.read_csv("student-mat.csv", sep=";")
por_df = pd.read_csv("student-por.csv", sep=";")

print("Math dataset shape:", mat_df.shape)
print("Portuguese dataset shape:", por_df.shape)
```

Step 4 — Merge the Datasets

```python
df = pd.concat([mat_df, por_df], axis=0, ignore_index=True)

print("Merged dataset shape:", df.shape)
df.head()
```

Step 5 — Validate Schema

```python
print("Columns in dataset:\n")
print(df.columns)

print("\nData Types:\n")
print(df.dtypes)
```

Step 6 — Check Missing Values

```python
missing_values = df.isnull().sum()
print("\nMissing values per column:\n", missing_values)

print("\nTotal Missing Values:", missing_values.sum())
```

Step 7 — Identify Feature Types

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

print("Categorical Features:", categorical_cols)
print("\nNumeric Features:", numeric_cols)
```

Step 8 — Identify Ordinal Features (integer features are actually ordinal scales (0–5 etc.))

```python
ordinal_cols = [
    'Medu','Fedu','traveltime','studytime','failures',
    'famrel','freetime','goout','Dalc','Walc','health'
]

print("Ordinal Features:", ordinal_cols)
```

Step 9 — Verify Target Variable (G3)

```python
print("Target variable range:")

print("Min G3:", df['G3'].min())
print("Max G3:", df['G3'].max())

# Check invalid values
invalid_g3 = df[(df['G3'] < 0) | (df['G3'] > 20)]
print("Invalid G3 rows:", len(invalid_g3))
```

Step 10 — Remove Leakage Features (G1 & G2) - these must be excluded from training features.

```python
df_clean = df.drop(columns=['G1','G2'])

print("Columns after removing leakage features:")
print(df_clean.columns)
```

Step 11 — Final Dataset Validation

```python
print("Final dataset shape:", df_clean.shape)

df_clean.info()
```

Step 12 — Save Clean Versioned Dataset

```python
df_clean.to_csv("student_performance_clean_v1.csv", index=False)

print("Clean dataset saved successfully")
```
